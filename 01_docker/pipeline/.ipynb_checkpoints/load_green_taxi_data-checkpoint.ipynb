```python
import click
import pandas as pd
from sqlalchemy import create_engine
from tqdm.auto import tqdm
from pathlib import Path
```

```python
# project root is current working directory (where the kernel was started)
project_root = Path.cwd()

data_path = project_root / "raw_data" / "green_tripdata_2025-11.parquet"
# /pipeline/raw_data/green_tripdata_2025-11.parquet

print(data_path)
# url = 'https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_/green_tripdata_2025-11.parquet'
```

```python
df = pd.read_parquet(data_path)
# len(df)
```

```python
# Create the connection to postgres
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
df.to_sql(
    name='green_taxi_trips',
    con=engine,
    if_exists='replace'
)
```
